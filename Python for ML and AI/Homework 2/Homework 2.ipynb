**5.	Read about Pearson correlation coefficient and then find the correlation between the columns: 'share_population_in_metro_areas', 'median_household_income', and 'non_white_white_poverty_ratio'**

```python
'''
Source: https://realpython.com/numpy-scipy-pandas-correlation-python/

The Pearson (product-moment) correlation coefficient is a measure of the linear relationship between two features
(but can consider more than two features). It’s the ratio of the covariance of feature 1 and feaurer 2 to the product 
of their standard deviations. It’s often denoted with the letter r.

- The Pearson correlation coefficient can take on any real value in the range −1 ≤ r ≤ 1.
- The value r > 0 indicates positive correlation between the features.
- The value r = 0 corresponds to the case when features are independent.
- The value r < 0 indicates negative correlation between features.
'''

metro = crimes['share_population_in_metro_areas']
income = crimes['median_household_income']
poverty = crimes['non_white_white_poverty_ratio']

metro_income_poverty = pd.DataFrame({'share_population_in_metro_areas': metro, 'median_household_income': income, 'non_white_white_poverty_ratio': poverty})

correlation_matrix = metro_income_poverty.corr()
correlation_matrix
```

### Problem 2
#### Calculating the value of pi using the Monte Carlo method.
Read the document in the link: https://academo.org/demos/estimating-pi-monte-carlo/ and calculate the value of pi.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
total_num_points = 1000000
points = np.random.uniform(-0.5, 0.5, (total_num_points, 2))   # Generate an array of (x, y) coordinates within [-0.5, 0.5]
print(points)
```

```python
x = points[:,0]   # Isolate x values of the points in a column
y = points[:,1]   # Isolate y values of the points in a column
print(x, y)
```

```python
distances = np.sqrt(x**2 + y**2)   # Perform distance calculation on all pairs of x's and y's
print(distances)
```

```python
points_in_circle = points[distances <= 0.5]   # Find points within radius of circle (0.5)
print(points_in_circle)
```

```python
pi = 4 * (len(points_in_circle) / total_num_points)   # pi ~ 4*(N-inside/N-total)
print(pi)
```

```python
x_in = points_in_circle[ : ,0]   # Isolate x values of the points inside the circle in a column
y_in = points_in_circle[ : ,1]   # Isolate y values of the points inside the circle in a column
plt.scatter(x_in, y_in)   # Make a scatter plot of all points inside the circle
plt.show()
```
