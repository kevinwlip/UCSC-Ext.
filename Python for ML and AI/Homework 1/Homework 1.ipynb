### Problem 1
Generate 100 x and y coordinates where x = [0, 1] and y = 3x^3+23x^2+0.25x+2.9.<br />
Then store the x and y coordinates to a csv file.<br />
Ensure that there are only 4 digits after decimal point in the csv file.

```python
import numpy as np
import pandas as pd
```

```python
x = np.random.rand(100)
print(x)
```

```python
y = 3*x**3 + 23*x**2 + 0.25*x + 2.9
print(y)
```

```python
x_and_y = {'X': x, 'Y': y}
df = pd.DataFrame(x_and_y).round(4)
print(df)
```

```python
df.to_csv('data.csv', float_format='%.4f')
```

#### Locality sensitive hashing (LSH)
Read the document:<br />

> https://towardsdatascience.com/fast-near-duplicate-image-search-using-locality-sensitive-hashing-d4c16058efcb<br />

and implement locality sensitive hashing (LSH).<br />
You need to test only two hyperplanes x = 0.5 and y = 0.6.<br />
Begin by generating 100 random x, y coordinates in the range of [0, 1].<br />
Then test each of the coordinates whether they lie on either side of the hyperplane.<br />
Then print the x and y coordinate and also the boolean corresponding to the test for either hyperplane.

```python
import numpy as np
```

```python
x = np.random.rand(100)
print(x)
```

```python
y = np.random.rand(100)
print(y)
```

```python
coordinates = list(zip(x,y))
print(coordinates)
```

```python
planes = []
temp = []
for c in coordinates:
    if c[0] > 0.5:
        temp.append('+')
    elif c[0] < 0.5:
        temp.append('-')
    else:
        temp.append('=')
        
    if c[1] > 0.6:
        temp.append('+')
    elif c[1] < 0.6:
        temp.append('-')
    else:
        temp.append('=')
    planes.append(tuple(temp))
    temp.clear()

print(coordinates)
print()
print(planes)
```
